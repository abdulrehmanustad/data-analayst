```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

rundata=pd.read_csv('ICC Test Bat 3001.csv').head(100)
#TOP 10 RUNSCORER IN TEST HISTORY
top_10=rundata.loc[:,['Player','Runs']]
top_10['Runs']=pd.to_numeric(top_10['Runs'])
top_10=top_10.sort_values(by='Runs',ascending=False).head(10).reset_index(drop=True)
print(top_10)
plt.figure(figsize=(10,6))
sns.barplot(x='Player',y='Runs',data=top_10)

plt.xlabel('Player')

plt.ylabel('Runs')
plt.title('Top 10 Run Scorers')
plt.xticks(rotation=45)
plt.show()
```

```python
# MOST 100 BY A INDIVIDUAL
most_100=rundata.loc[:,['Player','100']]
most_100['100']=pd.to_numeric(most_100['100'])
most_100=most_100.sort_values(by='100',ascending=False).head(10)
print(most_100)

plt.figure(figsize=(10,6))
sns.barplot(x='Player',y='100',data=most_100)
plt.xlabel('Player')
plt.ylabel('100s')
plt.title('Top 10 Scorers with 100s')
plt.xticks(rotation=45)
plt.show()
```

```python
# MOST 50 BY A INDIVIDUAL
most_50=rundata.loc[:,['Player','50']]
most_50['50']=pd.to_numeric(most_50['50'])
most_50=most_50.sort_values(by='50',ascending=False).head(10)
print(most_50)

plt.figure(figsize=(10,6))
sns.barplot(x='Player',y='50',data=most_50,color='red')
plt.xlabel('Player')
plt.ylabel('50s')
plt.title('Top 10 Scorers with 50s')
plt.xticks(rotation=45)
plt.show()
```

```python
# HIGHEST INDIVIDUAL SCORE BY A PLAYER
high_score=rundata.loc[:,['Player','HS']]
high_score['HS']=pd.to_numeric(high_score['HS'], errors='coerce')
high_score=high_score.sort_values(by='HS',ascending=False).head(10)
print(high_score)
plt.figure(figsize=(10,6))
sns.barplot(x='Player',y='HS',data=high_score,color='green')
plt.xlabel('Player')
plt.ylabel('Highest Score')
plt.title('Top 10 Highest Scores')
plt.xticks(rotation=45)
plt.show()
```
