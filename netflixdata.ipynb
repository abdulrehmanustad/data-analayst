```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

net_data=pd.read_csv('netflixdata.csv',engine='python')
print(net_data.head())
most_popular=net_data.loc[:,['Title','Popularity']]
most_popuar=pd.to_numeric(most_popular['Popularity'],errors='coerce')
most_popular=most_popular.sort_values(by=['Popularity'],ascending=False)
print(most_popular.head(10))
plt.figure(figsize=(10,5))
sns.barplot(x='Title',y='Popularity',data=most_popular.head(10),color='green')
plt.xlabel('Title')
plt.ylabel('Popularity')
plt.title('Top 10 Most Popular Movies on Netflix')
plt.xticks(rotation=30)
plt.show()

```

```python
most_voted=net_data.loc[:,['Title','Vote_Count']]
most_voted['Vote_Count']=pd.to_numeric(most_voted['Vote_Count'],errors='coerce')
most_voted=most_voted.sort_values(by=['Vote_Count'],ascending=False)
print(most_voted.head(10))
plt.figure(figsize=(10,5))
sns.barplot(x='Title',y='Vote_Count',data=most_voted.head(10),color='red')
plt.xlabel('Title')
plt.ylabel('Vote Count')
plt.title('Top 10 Most Voted Movies on Netflix')
plt.xticks(rotation=30)
plt.show()
```

```python
most_rated=net_data.loc[:,['Title','Vote_Average']]
most_rated['Vote_Average']=pd.to_numeric(most_rated['Vote_Average'],errors='coerce')
most_rated=most_rated.sort_values(by=['Vote_Average'],ascending=False)
print(most_rated.head(10))
plt.figure(figsize=(10,5))
sns.barplot(x='Title',y='Vote_Average',data=most_rated.head(10))
plt.xlabel('Title')
plt.ylabel('Vote Average')
plt.title('Top 10 Most Rated Movies on Netflix')
plt.xticks(rotation=30)
plt.show()

```
